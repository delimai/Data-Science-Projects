# Cleaning Data 

This time, I am going to clean this dataset until its ready for analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Hyper Cars 2019.csv')
```

```python
df
```

```python
df.columns
```

### Renaming the columns

```python
df.rename(columns={'     Car Name': 'car_serie', 'Displacement ': 'displacement', 'Enginee': 'engine', 
                   '          hp': 'hp', '   Transmission': 'transmission', 
                   '    Toppp-speeed': 'top_speed', '           Cost': 'cost'}, inplace=True)
```

```python
df.info()
```

### Removing the ',' from the cost table

```python
df['cost'] = df['cost'].apply(lambda x: x.replace(",","").replace(' ', ''))
```

```python
df['cost']
```

```python
df['cost'] = df['cost'].astype('float')
```

```python
df.info()
```

```python
df
```

```python
car_name = df.pop('car_serie')
df.insert(0, 'car_name', car_name)
```

```python
df
```

```python
df['displacement'] = df['displacement'].replace('Hello', '4.6')
```

```python
df['displacement'] = df['displacement'].astype('float').round(2)
```

```python
df['car_name'] = df['car_name'].replace('McLaren Sennaaaa', 'McLaren Sena')
df['car_name'] = df['car_name'].replace('Koenigseeeggg Agera', 'Koenigsegg Agera')
df['car_name'] = df['car_name'].replace('Pagaaani Huayra BC', 'Pagani Huayra BC')
```

```python
df
```

```python
df['engine'] = df['engine'].replace('V8987654', 'V8')
df['engine'] = df['engine'].replace('__', 'V12')
```

```python
df.isnull().sum()
```

```python
df[df['hp'].isnull()]
```

```python
#df['hp'] = df['hp'].fillna(599)
```

```python
df.at[5, 'hp'] = 599
df.at[8, 'hp'] = 1160
```

```python
#filling NaN in trnasmission with the mean of that column
df['transmission'] = df['transmission'].fillna(df['transmission'].mean())
```

```python
df.at[6, 'top_speed'] = 350
```

```python
df['top_speed'] = df['top_speed'].round(2)
```
